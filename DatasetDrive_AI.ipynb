```python
def make_trainData(name,DIR):
  for img in os.listdir(DIR):
    label = label_img(img,name)
    path = os.path.join(DIR,img)
    img = cv2.imread(path)
    img = cv2.resize(img,(Img_size,Img_size))
    X.append(np.array(img))
    W.append(str(label))
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
Label = LabelEncoder()
Y = Label.fit_transform(W)
Y=to_categorical(Y,4)
X=np.array(X)
X=X/255.0
print(type(X))
print(type(Y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

x_train=np.array(x_train).reshape(465,10000*3) 
x_test=np.array(x_test).reshape(155,10000*3) 
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model = Sequential()
model.add(tf.keras.layers.Dense(units=145,activation="relu"))
model.add(tf.keras.layers.Dense(units=145,activation="relu"))
model.add(tf.keras.layers.Dense(units=145,activation="relu"))

model.add(tf.keras.layers.Dense(units=4,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=32,epochs = 50)
```

```python
ModelLoss, ModelAccuracy = model.evaluate(x_test, y_test)

print('Test Loss is {}'.format(ModelLoss))
print('Test Accuracy is {}'.format(round(ModelAccuracy*100) ),'%')
```
